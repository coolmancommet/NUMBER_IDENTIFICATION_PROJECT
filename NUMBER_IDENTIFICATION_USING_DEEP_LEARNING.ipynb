# DATA prepration 
#### the given data is in the form of pandas dataframe and having 784 columns 
(excluding the label column)

now these columns are the pixel intensities of the 28 X 28 pixel image where 0 means white(or full intensity) and 255 means complete black(or 0 intensity)

*to convert these numbers into 28 X 28 n-dimensional array we will use the function* 

```
data_prep()
```
*this function takes input of raw datafrae with 784 columns and give you two arrays-*

*1- list of 28 x 28 n-dimentional array*

*2- the labels of the following*





````
num_classes=10 #number of classes in the data set
img_row=img_col=28 #no. of rows and columns on the image
````

````
def data_prep(raw_data):
    raw=raw_data.to_numpy()
    y=raw[:,0]
    out_y=keras.utils.to_categorical(y,num_classes)
    x = raw[:,1:]
    num_images = raw.shape[0]
    out_x = x.reshape(num_images, img_row, img_col, 1)
    out_x = out_x / 255
    return out_x, out_y
````

````
x,y=data_prep(train)
````

# MODEL PREPRATION 
 
---

![alt text]()
In this section we will create a new model from scratch now you can increase or decrease the no. of the layers in the model but try to keep the model as simple as possible as the more no. of layes you going to add the more your model will get complex and it will be difficult to handle ,it will take much more time ,their can be a possibility of overfitting etc.

*now here i used 4 convolution layers(1 input layer and 3 hidden) with 3 X 3 filter and without any strides or pooling layer after that to flatten my output of the third layer I used flatten layer after that a fully connected layer and for calsification again a dense layer*
 

````
my_new_model=Sequential()
my_new_model.add(Conv2D(28,kernel_size=(3,3),activation='relu'))

my_new_model.add(Conv2D(20,kernel_size=(3,3),activation='relu'))
my_new_model.add(Conv2D(18,kernel_size=(3,3),activation='relu'))
my_new_model.add(Conv2D(15,kernel_size=(3,3),activation='relu'))

my_new_model.add(Flatten())
my_new_model.add(Dense(100,activation='relu'))
my_new_model.add(Dense(10,activation='softmax'))
````

````
my_new_model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
````

````
my_new_model.fit(x,y,batch_size=100,epochs=40,validation_split=0.2)
````

# DATA PREPRATION FOR THE CUSTOM IMAGES BY USER

````
def imageprepare(argv):
    """
    This function returns the pixel values.
    The imput is a png file location.
    """
    im = Image.open(argv).convert('L')
    width = float(im.size[0])
    height = float(im.size[1])
    newImage = Image.new('L', (28, 28), (255))  # creates white canvas of 28x28 pixels

    if width > height:  # check which dimension is bigger
        # Width is bigger. Width becomes 20 pixels.
        nheight = int(round((20.0 / width * height), 0))  # resize height according to ratio width
        if (nheight == 0):  # rare case but minimum is 1 pixel
            nheight = 1
            # resize and sharpen
        img = im.resize((20, nheight), Image.ANTIALIAS).filter(ImageFilter.SHARPEN)
        wtop = int(round(((28 - nheight) / 2), 0))  # calculate horizontal position
        newImage.paste(img, (4, wtop))  # paste resized image on white canvas
    else:
        # Height is bigger. Heigth becomes 20 pixels.
        nwidth = int(round((20.0 / height * width), 0))  # resize width according to ratio height
        if (nwidth == 0):  # rare case but minimum is 1 pixel
            nwidth = 1
            # resize and sharpen
        img = im.resize((nwidth, 20), Image.ANTIALIAS).filter(ImageFilter.SHARPEN)
        wleft = int(round(((28 - nwidth) / 2), 0))  # caculate vertical pozition
        newImage.paste(img, (wleft, 4))  # paste resized image on white canvas

    # newImage.save("sample.png

    tv = list(newImage.getdata())  # get pixel values

    # normalize pixels to 0 and 1. 0 is pure white, 1 is pure black.
    tva = [(255 - x) for x in tv]
    return pd.DataFrame(data=[tva],index=None,columns=['pixel_{}'.format(n+1) for n in range(0,784)])

def imgprrr(folder_name):
  """
  this function takes the input of the folder path 
  and return the dataframe with pixel intensities of the images in the order they are in the folder
  """
  content=os.listdir(folder_name)
  ret_datafr=pd.DataFrame(index=None,columns=['pixel_{}'.format(n+1) for n in range(0,784)])
  for image_name in content:
    temp_datafr=imageprepare(folder_name+'/'+image_name)
    ret_datafr=pd.concat([ret_datafr,temp_datafr])
  return ret_datafr
````
